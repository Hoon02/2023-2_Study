# What I learned

```python
import matplotlib.pyplot as plt
import numpy as np

l1 = 1
l2 = 2
l3 = -3
v1 = np.array([4, 5, 1])
v2 = np.array([-4, 0 , -4])
v3 = np.array([1, 3, 2])
l1*v1 + l2*v2 + l3*v3
```

# Exercise 2-1

```python
sl = [l1, l2, l3]
vl = [v1, v2, v3]
linCombo = np.zeros(len(v1))
for s,v in zip(sl, vl):
    linCombo += s*v

linCombo
```

# Exercise 2-2

```python
sl = [l1, l2, l3 , 5]
vl = [v1, v2, v3]
linCombo = np.zeros(len(v1))
for i in range(len(sl)):
    linCombo += sl[i]*vl[i]

linCombo
# for문 사용시 확장성 떨어짐
```

# Exercise 2-3

```python
v1 = np.array([1, 3])
xlim = [-4, 4]

sl = np.random.uniform(low=-4, high=4, size=100)

plt.figure(figsize=(6,6))

for s in sl:
    
    p = v1*s
    plt.plot(p[0], p[1], 'ko')
    
plt.xlim(xlim)
plt.ylim(xlim)
plt.grid()
plt.show()
```

```python
import plotly.graph_objects as go

v1 = np.array([3, 5, 1])
v2 = np.array([0, 2, 2])

sl = np.random.uniform(low=xlim[0], high=xlim[1], size=(150,2))

points = np.zeros((150, 3))
for i in range(len(sl)):
    points[i,:] = v1*sl[i,0] + v2*sl[i, 1]
    

fig = go.Figure(data=[go.Scatter3d(x=points[:,0], y=points[:,1], z=points[:,2], 
                                    mode='markers', marker=dict(size=6,color='black') )])

fig.update_layout(margin=dict(l=0,r=0,b=0,t=0))
plt.savefig('Figure_02_07b.png',dpi=300)
fig.show()
```
